```python
import yaml
import pandas as pd
import os

# Step 1: Load YAML
with open("../../config/merging.yaml", "r") as f:
    config = yaml.safe_load(f)

flow_file = config["OD"]['flow']
msa_file = config["OD"]["msa"]
```

```python
path = '../../data/'
```

```python
flow_data = pd.read_csv(path+flow_file)
msa_data = pd.read_csv(path+msa_file)
```

## Merge Flow(Left) and MSA Attributes (Right)

```python
# Merge msa_data on Origin_Code
flow_with_origin = flow_data.merge(
    msa_data,
    left_on="Origin_Code",                # use Origin_Code from flow_data
    right_on="Crosswalk2023_CBSA Code",   # match with CBSA code from msa_data
    how="left",                           # keep all flows, bring in matching MSA info
    suffixes=("", "_drop")                # prevent duplicate column names
).rename(lambda x: f"Origin_{x}" if x in msa_data.columns else x, axis=1)

# Drop any duplicate columns created during merge
flow_with_origin = flow_with_origin.loc[:, ~flow_with_origin.columns.str.endswith("_drop")]

# Merge msa_data on Destination_Code
flow_final = flow_with_origin.merge(
    msa_data,
    left_on="Destination_Code",           # use Destination_Code from flow_data
    right_on="Crosswalk2023_CBSA Code",   # match with CBSA code from msa_data
    how="left",
    suffixes=("", "_drop")
).rename(lambda x: f"Destination_{x}" if x in msa_data.columns else x, axis=1)

# Drop duplicates again
flow_final = flow_final.loc[:, ~flow_final.columns.str.endswith("_drop")]

# Show final result with both Origin_ and Destination_ prefixed attributes
flow_final.head()
```

```python
flow_final.to_csv(f'../../data/final/flow_data_final.csv', index=False)
```

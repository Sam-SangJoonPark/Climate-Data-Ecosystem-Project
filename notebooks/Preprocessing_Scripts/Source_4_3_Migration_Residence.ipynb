## Leave only necessary columns - Residence information

```python
cols = df.columns[:15].drop(df.columns[1])
df = df[cols]
```

### Sanity Check

### Population =  Nonmovers + Movers (Within MSA, Different MSA, Elsewhere in the U.S. + Abroad)

```python
df['sum_check'] = df['Metropolitan Statistical Area of Current Residence_Nonmovers_Estimate'] + df['Metropolitan Statistical Area of Current Residence_Movers within Same Metropolitan Statistical Area_Estimate'] + df['Metropolitan Statistical Area of Current Residence_Movers from Different Metropolitan Statistical Area2_Estimate'] + df['Metropolitan Statistical Area of Current Residence_Movers from Elsewhere in the U.S. or Puerto Rico_Estimate'] + df['Metropolitan Statistical Area of Current Residence_Movers from Abroad3_Estimate']
```

```python
df[['Metropolitan Statistical Area of Current Residence_Population 1 Year and Over_Estimate', 'sum_check']].sample(10)
```

```python
del df['sum_check']
```

```python
cols = ['Current_Residence_Code', 'MSA_Name',
        'Current_Residence_Population', 'Current_Residence_Population_MOE',
        'Current_Residence_Nonmovers', 'Current_Residence_Nonmovers_MOE',
        'Current_Residence_Movers_within_MSA', 'Current_Residence_Movers_within_MSA_MOE',
        'Current_Residence_Movers_different_MSA', 'Current_Residence_Movers_different_MSA_MOE',
        'Current_Residence_elsewhere_PR', 'Current_Residence_elsewhere_PR_MOE',
        'Current_Residence_Movers_abroad', 'Current_Residence_Movers_abroad_MOE']
```

```python
len(df.columns), len(cols)
```

## Leave only essential rows

```python
df_msa_level = (
    df.groupby(["Current_Residence_Code", "MSA_Name"], as_index=False)
      .first()
)
```

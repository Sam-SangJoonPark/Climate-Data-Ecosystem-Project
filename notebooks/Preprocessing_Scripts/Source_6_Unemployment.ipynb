## Unemployment data preprocess (FIPS Code to CBSA Code)

```python
def add_fips(df):
    """
    Create a 5-digit FIPS code by combining State FIPS Code and County FIPS Code.
    - State FIPS Code: keep as int → str (no zero padding)
    - County FIPS Code: 3-digit zero padding
    """
    df = df.copy()
    
    df["State FIPS Code"] = df["State FIPS Code"].astype(int).astype(str)
    df["County FIPS Code"] = df["County FIPS Code"].astype(int).astype(str).str.zfill(3)
    df["FIPS"] = df["State FIPS Code"] + df["County FIPS Code"]
    
    return df
```

```python
# Put all DataFrames into a list
dfs = [df0, df1, df2, df3, df4, df5, df6, df7, df8, df9]

# Apply preprocessing function to each
dfs_processed = [add_fips(d) for d in dfs]

# Optionally unpack back into variables
df0, df1, df2, df3, df4, df5, df6, df7, df8, df9 = dfs_processed
```

## Match to MSA Level

```python
version = 2023
crosswalk_data_path = f'../../data/interim/data2_county_level_{version}.csv'

crosswalk_data = pd.read_csv(crosswalk_data_path)

crosswalk_data.head()
```

```python
unemployment_panel
```

```python
# Ensure both keys are strings
unemployment_panel["FIPS"] = unemployment_panel["FIPS"].astype(str)
crosswalk_data["Crosswalk2023_FIPS_Key"] = crosswalk_data["Crosswalk2023_FIPS_Key"].astype(str)

unemployment_with_msa = unemployment_panel.merge(
    crosswalk_data,
    how="left",
    left_on="FIPS",
    right_on="Crosswalk2023_FIPS_Key"
)
```

```python
unemployment_with_msa.columns
```

```python
unemployment_with_msa
```

```python
# Group by Year × MSA
unemployment_summary = (
    unemployment_with_msa
    .groupby(["Year", "Crosswalk2023_CBSA Code"], as_index=False)
    .agg({
        "Labor Force": "sum",
        "Employed": "sum",
        "Unemployed": "sum"
    })
)
```

```python
unemployment_summary
```

```python
unemployment_summary.isnull().sum()
```

```python
# Pivot: wide format by Year
unemp_final = unemployment_summary.pivot_table(
    index="Crosswalk2023_CBSA Code",
    columns="Year",
    values=["Labor Force", "Employed", "Unemployed"],
    aggfunc="sum"
)

# Flatten columns with desired order: Labor Force → Employed → Unemployed
ordered_vars = ["Labor Force", "Employed", "Unemployed"]
ordered_cols = []

for var in ordered_vars:
    years = sorted(unemp_final[var].columns)  # ensure years are sorted
    ordered_cols.extend([f"{var.replace(' ', '_')}_{int(y)}" for y in years])

# Rename columns and reorder
unemp_final.columns = [f"{col[0].replace(' ', '_')}_{int(col[1])}" for col in unemp_final.columns]
unemp_final = unemp_final[ordered_cols].reset_index()
```

## Climate Risk Related Indices

```python
# total_population_equiv
population_equivalence_calculation = 11_600_000
final_df['total_population_equiv'] = final_df['NRI_POPULATION'] * population_equivalence_calculation

# exposure_pop_total (exposure population equivalence)
exppe_cols = [col for col in final_df.columns if "_EXPPE" in col]
final_df["exposure_pop_total"] = final_df[exppe_cols].sum(axis=1)

# pct_pop_exposed
final_df['pct_pop_exposed'] = 100 * final_df["exposure_pop_total"] / final_df['total_population_equiv']

# loss_pop_total (loss population equivalnce)
pe_cols = [col for col in final_df.columns if "_EALPE" in col]
final_df["loss_pop_total"] = final_df[pe_cols].sum(axis=1)

# pct_pop_loss
final_df['pct_pop_loss'] = 100 * final_df["loss_pop_total"] / final_df['exposure_pop_total']

# building_equiv
building_equivalence_calculation = 17
final_df['building_equiv'] = final_df['NRI_BUILDVALUE'] * building_equivalence_calculation

# exposure_bldg_total
expb_cols = [col for col in final_df.columns if "_EXPB" in col]
final_df["exposure_bldg_total"] = final_df[expb_cols].sum(axis=1)

# pct_bldg_exposed
final_df['pct_bldg_exposed'] = 100 * final_df["exposure_bldg_total"] / final_df['building_equiv']

# loss_bldg_total
ealb_cols = [col for col in final_df.columns if "_EALB" in col]
final_df["loss_bldg_total"] = final_df[ealb_cols].sum(axis=1)

# pct_bldg_loss
final_df['pct_bldg_loss'] = 100 * final_df["loss_bldg_total"] / final_df['exposure_bldg_total']

# loss_per_capita_pe
final_df['loss_per_capita_pe'] = final_df["loss_pop_total"] / final_df['NRI_POPULATION']

# loss_per_capita_bldg
final_df['loss_per_capita_bldg'] = final_df["loss_bldg_total"] / final_df['NRI_POPULATION']


## Total (PE + BLDG)
# exposure_total
final_df['climate_exposure_total'] = final_df["exposure_pop_total"] + final_df["exposure_bldg_total"]

# loss_total
final_df['climate_loss_total'] = final_df["loss_pop_total"] + final_df["loss_bldg_total"]
```

## Unemployment Indices

```python
for year in range(2015, 2025):  # 2015 ~ 2024
    unemp_col = f"Unemployed_{year}"
    labor_col = f"Labor_Force_{year}"
    rate_col = f"Unemployment_Rate_{year}"
    
    if unemp_col in final_df.columns and labor_col in final_df.columns:
        final_df[rate_col] = final_df[unemp_col] / final_df[labor_col] * 100
```

```python
# 2016~2020 Sum of Unemployed
unemp_cols = [f"Unemployed_{year}" for year in range(2016, 2021)]
final_df["Unemployed_2016_2020_sum"] = final_df[unemp_cols].sum(axis=1, skipna=True)

# 2016~2020 Sum of Labor Force
lf_cols = [f"Labor_Force_{year}" for year in range(2016, 2021)]
final_df["LaborForce_2016_2020_sum"] = final_df[lf_cols].sum(axis=1, skipna=True)

## average unemployment rate
final_df['avg_unemployment_rate_16_20'] = 100 * final_df["Unemployed_2016_2020_sum"] / final_df["LaborForce_2016_2020_sum"]

# check results
final_df[["Unemployed_2016_2020_sum", "LaborForce_2016_2020_sum", "avg_unemployment_rate_16_20"]].head()
```

## Housing PI

```python
# 2016 ~ 2020 columns
cols = ["HousingPI25_2016", "HousingPI25_2017", "HousingPI25_2018", "HousingPI25_2019", "HousingPI25_2020"]

# Mean (ignore NaN)
final_df["HousingPI25_2016_2020_mean"] = final_df[cols].mean(axis=1)

# check results
final_df[["HousingPI25_2016_2020_mean"]].head()
```
